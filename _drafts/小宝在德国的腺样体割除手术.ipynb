小儿从刚出生开始，睡觉就打鼾，长大一点还出现张嘴呼吸的现象。我们作为年轻父母，非常地无知，刚开始还觉得挺有意思，这么小的baby也打鼾。所幸后来也慢慢意识到，这肯定不太好，于是前几次体检时，都会提醒儿医睡觉打呼噜的事。医生一般解释说样腺体增大对小宝宝来说比较常见，通常情况下，会自动缩小，打呼噜和张嘴呼吸的情况也会慢慢消失。时间飞快，小家伙很快就两岁多了，但每天晚上在隔壁房间就能听到他的鼾声，没打鼾时，就是仰头望天的姿势，张嘴呼吸。有一次，偶然看到知乎上的[这个](https://www.zhihu.com/question/24972857)，被吓得不轻，赶紧约了耳鼻喉科的医生，看专业的医生怎么说。

### 先看耳鼻喉科
在德国看耳鼻喉科（HNO Arzt），一般不需要家庭医生(Hausarzt)或者儿科医生(Kinderarzt)的转诊单(Ueberweisungschein)。在google review上找了个好评较高的老医生Brueckemann，连约都不用约，就直奔他那儿了。老爷子看看喉，看看耳朵，半分钟不到，就跟我说，这个小娃样腺体增大，现在有两个方案，要么再等两年，看是否会好些，要么去医院手术，把样腺体割除。我就是抱着要手术的打算去的，直接让他帮我开转诊单，咱去医院把它给割了。

在德国看医生，发现跟国内一个很大的不同就是，这里很少用X光，CT，或者核磁之类的检查手段。比如，这个耳鼻喉科医生，看看耳朵看看喉咙，就让你做手术了，国内或许需要更多一些检查手段才会给诊断结论。这或许是为了节省医疗资源，或许是私立诊所配备这些诊断设备的成本过高，或者其他啥。

### 看医院的耳鼻喉科
我们转到了EVK (Evanglisch Krankenhaus)。当然，医院的科室也必须先确认下耳鼻喉科诊所的诊断结论，所以跟医院的第一个Termin就是再检查一遍，看是否需要手术。这基本上是重复老爷子的检查，然后说了下手术的大概过程，就是一大早过来，然后全身麻醉，真正手续只需要一两分钟，做完后在病房观察几个小时，一切正常后当天就可以回家。检查完后，就跟医院的耳鼻喉科住院部约手术时间和床位。由于是小娃娃的手术，床位安排在儿童住院部。回家后过了两天，医院又寄了一封信，详细说明手术时间和地点，并简要说明了注意事项。

### 手术过程
整个手术需要约两天，第一天一早带着娃去医院做预检查，看第二天是否能够做手术。第一天闹了个笑话，凭着半桶水的德语，我听错一句话，以为就是填一填资料然后跟医生简单聊聊，不需要带娃前往。哪想填完资料后，进到第一个医生的办公室，她一脸惊奇的看着我，你不带娃来，我给谁看。。。于是赶紧去幼儿园把小家伙拎出来了。第一个医生就是做个简单的体检，看是否感冒。第二个医生，是耳鼻喉科另外一个医生再检查一遍，然后详细说明手术过程，解释手术可能的风险。第三个医生是跟麻醉师聊，问孩子是否有病史，有过敏源之类的问题，然后说明麻醉的过程和风险。此处无需多言。

第一天的傍晚，让小娃吃好喝好，然后让他睡好。因为第二天凌晨开始，就不能喝水，不能吃东西，直到手术结束。后来发现这个其实对一个两岁多的娃来说是个很大的折磨和挑战。早上六点就从家出发，6点半到医院的耳鼻喉科住院部，当天还有另外一个三岁半的小朋友做扁桃体割除手术。等待的时候两个小朋友被安排在了同一个房间，有玩具，有书，可以舒缓小朋友的情绪。护士先给他们手上贴了麻醉测试片，大约七点半的时候，护士拿来了口服麻醉药。我家小宝乖乖喝完，然后我给读故事书，不出十分钟，他就迷迷糊糊的痴笑，口里还念念有词，药效上来了，不一会他就睡着了。大概八点左右，护士就把他推到手术室了。等到九点左右，护士跟我说，手术做完了，让我到观察室陪小宝。
